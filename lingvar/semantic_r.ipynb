```r
txt = read.table("rows.txt", 
    sep="\t", 
    header=FALSE, 
    comment.char="",
    quote="",
    stringsAsFactors=FALSE,
)
res = list()
for (i in 1:nrow(txt)) {
  s = strsplit(txt[i, ], ' ')[[1]]
  s2 = data.frame(
       Cues = rep(" ", length(s)),
       Outcomes = rep(" ", length(s)),
       Frequency = rep(1, length(s)), stringsAsFactors=FALSE)
  for (j in 1:length(s)) {
    s2[j,"Cues"] = paste(s[-j], sep="_", collapse="_")
    s2[j,"Outcomes"] = s[j]
  }
  res[[i]] = s2
  if (i %% 1000 == 0) {
      cat(".")
  }
}
res =do.call(rbind, res)
```
